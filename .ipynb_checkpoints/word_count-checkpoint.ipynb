```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
import csv
```

```python
# import files as lists
with open('good_summ.csv', 'r') as f:
    reader = csv.reader(f)
    good_summary = list(reader)

with open('poor_summ.csv', 'r') as f:
    reader = csv.reader(f)
    poor_summary = list(reader)
    
with open('good_reviewText.csv', 'r') as f:
    reader = csv.reader(f)
    good_reviewText = list(reader)

with open('poor_reviewText.csv', 'r') as f:
    reader = csv.reader(f)
    poor_reviewText = list(reader)
```

```python
# for good summary
gd_sum_wordcount = {}

for sentence in good_summary:
    quote = sentence[0].split()
    for word in quote:
        if word not in gd_sum_wordcount:
            gd_sum_wordcount[word] = 1
        else:
            gd_sum_wordcount[word] += 1
            
# gd_sum_wordcount = sorted(gd_sum_wordcount.items(), key=lambda kv: kv[1], reverse=True)

# for poor summary
poor_sum_wordcount = {}

for sentence in poor_summary:
    quote = sentence[0].split()
    for word in quote:
        if word not in poor_sum_wordcount:
            poor_sum_wordcount[word] = 1
        else:
            poor_sum_wordcount[word] += 1
            
# poor_sum_wordcount = sorted(poor_sum_wordcount.items(), key=lambda kv: kv[1], reverse=True)

# type(poor_sum_wordcount)
# poor_sum_wordcount
```

```python
# remove duplicates from other dictionary
def removeDuplicates( d1, d2):
    dict = {}
    for key,value in d1.items():
        if key not in d2:
            dict[key] = value
    return dict


gd_sum_results = removeDuplicates(gd_sum_wordcount, poor_sum_wordcount)
gd_sum_results = sorted(gd_sum_results.items(), key=lambda kv: kv[1], reverse=True)
# gd_sum_results

bd_sum_results = removeDuplicates(poor_sum_wordcount, gd_sum_wordcount)
bd_sum_results = sorted(bd_sum_results.items(), key=lambda kv: kv[1], reverse=True)
bd_sum_results
```
