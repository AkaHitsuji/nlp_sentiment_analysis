```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
import csv
```

```python
# import files as lists
def tolist(filename):
    with open(filename, 'r') as f:
        reader = csv.reader(f)
        return list(reader)
    
# summaries
good_summary = tolist('good_summ.csv')
poor_summary = tolist('poor_summ.csv')

# review text
good_reviewText = tolist('good_reviewText.csv')
poor_reviewText = tolist('poor_reviewText.csv')
```

```python
def get_wordcount(data):
    wordcount = {}

    for sentence in data:
        quote = sentence[0].split()
        for word in quote:
            if word not in wordcount:
                wordcount[word] = 1
            else:
                wordcount[word] += 1
                
    return wordcount

# based on summaries
gd_sum_wordcount = get_wordcount(good_summary)
poor_sum_wordcount = get_wordcount(poor_summary)

# based on review text
gd_rt_wordcount = get_wordcount(good_reviewText)
poor_rt_wordcount = get_wordcount(poor_reviewText)

gd_sum_wordcount = sorted(gd_sum_wordcount.items(), key=lambda kv: kv[1], reverse=True)
gd_sum_wordcount
```

```python
# remove duplicates from other dictionary
def removeDuplicates( d1, d2):
    dict = {}
    for key,value in d1.items():
        if key not in d2:
            dict[key] = value
    return dict

def get_results(wc1,wc2):
    results = removeDuplicates(wc1, wc2)
    return sorted(results.items(), key=lambda kv: kv[1], reverse=True)


gd_sum_results = get_results(gd_sum_wordcount, poor_sum_wordcount)
bd_sum_results = get_results(poor_sum_wordcount, gd_sum_wordcount)

gd_rt_results = get_results(gd_rt_wordcount,poor_rt_wordcount)
bd_rt_results = get_results(poor_rt_wordcount,gd_rt_wordcount)
```
